```python
import pandas as pd
import numpy as np

df = pd.read_csv("hotel_bookings.csv")

df.head()

```

```python
# Tamaño del dataset
filas, columnas = df.shape

print("Cantidad de filas:", filas)
print("Cantidad de columnas:", columnas)
print("Cantidad total de celdas:", filas * columnas)

# Primeras filas
display(df.head())

# Información de las columnas
df.info()

# Valores faltantes
faltantes = pd.DataFrame({
    "Cantidad": df.isnull().sum(),
    "Porcentaje": (df.isnull().sum() / len(df)) * 100
})

faltantes = faltantes[faltantes["Cantidad"] > 0]
display(faltantes.sort_values("Cantidad", ascending=False))

# Duplicados
print("Filas duplicadas:", df.duplicated().sum())
```

```python
df_original = df.copy()
df_limpio = df.copy()

print("Copia creada correctamente")
print("Filas originales:", df_limpio.shape[0])
print("Columnas originales:", df_limpio.shape[1])
```

```python
duplicados_antes = df_limpio.duplicated().sum()

df_limpio = df_limpio.drop_duplicates().reset_index(drop=True)

print("Duplicados encontrados:", duplicados_antes)
print("Filas después de eliminar duplicados:", df_limpio.shape[0])
```

```python
#Eliminar la columna "company" ya que tiene mas del 94% de valores faltantes, imputarala no seria confiable
df_limpio = df_limpio.drop(columns=["company"])

print("Columna 'company' eliminada")
print("Cantidad de columnas:", df_limpio.shape[1])
```

```python
#Tratar los valores faltantes
#Para children, usamos la mediana
df_limpio["children"] = df_limpio["children"].fillna(
    df_limpio["children"].median()
)
#Para country usamos, usamos la moda
df_limpio["country"] = df_limpio["country"].fillna(
    df_limpio["country"].mode()[0]
)
#Para agent, conviene usar 0, porque generalmente indica que la reserva no fue gestionada por un agente
df_limpio["agent"] = df_limpio["agent"].fillna(0)
#Comprobar el resultado
print("Valores faltantes restantes:")
print(df_limpio.isnull().sum()[df_limpio.isnull().sum() > 0])

print("\nDimensiones actuales:")
print(df_limpio.shape)

resumen_limpieza = pd.DataFrame({
    "Tipo de dato": df_limpio.dtypes,
    "Valores faltantes": df_limpio.isnull().sum()
})

display(resumen_limpieza)
```

```python
print(df_limpio["reservation_status"].value_counts())

display(
    df_limpio[
        ["is_canceled", "reservation_status", "reservation_status_date"]
    ].head(15)
)
```

```python
#Eliminacion de columnas con fugas de informacion
columnas_fuga = [
    "reservation_status",
    "reservation_status_date"
]

df_limpio = df_limpio.drop(columns=columnas_fuga)

print("Columnas eliminadas:", columnas_fuga)
print("Dimensiones actuales:", df_limpio.shape)
```

```python
reservas_sin_huespedes = df_limpio[
    (df_limpio["adults"] == 0) &
    (df_limpio["children"] == 0) &
    (df_limpio["babies"] == 0)
]

print("Reservas sin huéspedes:", len(reservas_sin_huespedes))

#Eliminar esas reservas sin huespedes
df_limpio = df_limpio[
    ~(
        (df_limpio["adults"] == 0) &
        (df_limpio["children"] == 0) &
        (df_limpio["babies"] == 0)
    )
].reset_index(drop=True)

print("Dimensiones después de eliminar reservas inválidas:")
print(df_limpio.shape)

#Recisamos el balance de la variable objetivo, saber cuantas reservas fueron canceladas
#y cuantas reservas no fueron canceladas
conteo_cancelaciones = df_limpio["is_canceled"].value_counts()

porcentaje_cancelaciones = (
    df_limpio["is_canceled"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

print("Cantidad:")
print(conteo_cancelaciones)

print("\nPorcentaje:")
print(porcentaje_cancelaciones)
print(df_limpio.shape)
print(df_limpio["is_canceled"].value_counts())
```

```python
#Separar la variable objetivo
#La columna para predecir es is_canceled

# Variable objetivo
objetivo = "is_canceled"

# X contiene las variables predictoras
X = df_limpio.drop(columns=[objetivo])

# y contiene la respuesta que queremos predecir
y = df_limpio[objetivo]

print("Dimensiones de X:", X.shape)
print("Dimensiones de y:", y.shape)
```

```python
#Para identificar columnas numericas y categoricas
columnas_numericas = X.select_dtypes(
    include=["int64", "float64"]
).columns.tolist()

columnas_categoricas = X.select_dtypes(
    include=["object", "category"]
).columns.tolist()

print("Columnas numéricas:", len(columnas_numericas))
print(columnas_numericas)

print("\nColumnas categóricas:", len(columnas_categoricas))
print(columnas_categoricas)
```

```python
#Para dividir entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)

#(stratify=y mantiene la misma proporción de reservas canceladas y no canceladas en entrenamiento y prueba.)

#Crear el procesador
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Escala las columnas numéricas
transformador_numerico = StandardScaler()

# Convierte las categorías en columnas binarias
transformador_categorico = OneHotEncoder(
    handle_unknown="ignore",
    sparse_output=True
)

# Une ambos procesos
preprocesador = ColumnTransformer(
    transformers=[
        (
            "numericas",
            transformador_numerico,
            columnas_numericas
        ),
        (
            "categoricas",
            transformador_categorico,
            columnas_categoricas
        )
    ]
)

print("Preprocesador creado correctamente")

#Transformar los datos
X_train_procesado = preprocesador.fit_transform(X_train)

X_test_procesado = preprocesador.transform(X_test)

print("Entrenamiento procesado:", X_train_procesado.shape)
print("Prueba procesada:", X_test_procesado.shape)
```

```python
#Agente 1 Normalizador
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer


class AgenteNormalizador:
    def __init__(self, columna_objetivo="is_canceled"):
        self.columna_objetivo = columna_objetivo
        self.preprocesador = None
        self.columnas_numericas = []
        self.columnas_categoricas = []

    def limpiar_datos(self, df):
        df_limpio = df.copy()

        # Eliminar filas duplicadas
        df_limpio = df_limpio.drop_duplicates()

        # Eliminar columnas con demasiados valores faltantes
        df_limpio = df_limpio.drop(
            columns=["company"],
            errors="ignore"
        )

        # Imputar valores faltantes
        if "children" in df_limpio.columns:
            df_limpio["children"] = df_limpio["children"].fillna(
                df_limpio["children"].median()
            )

        if "country" in df_limpio.columns:
            df_limpio["country"] = df_limpio["country"].fillna(
                df_limpio["country"].mode()[0]
            )

        if "agent" in df_limpio.columns:
            df_limpio["agent"] = df_limpio["agent"].fillna(0)

        # Eliminar columnas con fuga de información
        df_limpio = df_limpio.drop(
            columns=[
                "reservation_status",
                "reservation_status_date"
            ],
            errors="ignore"
        )

        # Eliminar reservas sin huéspedes
        columnas_huespedes = {"adults", "children", "babies"}

        if columnas_huespedes.issubset(df_limpio.columns):
            df_limpio = df_limpio[
                ~(
                    (df_limpio["adults"] == 0) &
                    (df_limpio["children"] == 0) &
                    (df_limpio["babies"] == 0)
                )
            ]

        return df_limpio.reset_index(drop=True)

    def preparar_datos(self, df):
        df_limpio = self.limpiar_datos(df)

        X = df_limpio.drop(columns=[self.columna_objetivo])
        y = df_limpio[self.columna_objetivo]

        self.columnas_numericas = X.select_dtypes(
            include=["int64", "float64"]
        ).columns.tolist()

        self.columnas_categoricas = X.select_dtypes(
            include=["object", "category"]
        ).columns.tolist()

        X_train, X_test, y_train, y_test = train_test_split(
            X,
            y,
            test_size=0.20,
            random_state=42,
            stratify=y
        )

        self.preprocesador = ColumnTransformer(
            transformers=[
                (
                    "numericas",
                    StandardScaler(),
                    self.columnas_numericas
                ),
                (
                    "categoricas",
                    OneHotEncoder(
                        handle_unknown="ignore",
                        sparse_output=True
                    ),
                    self.columnas_categoricas
                )
            ]
        )

        X_train_procesado = self.preprocesador.fit_transform(X_train)
        X_test_procesado = self.preprocesador.transform(X_test)

        return {
            "df_limpio": df_limpio,
            "X_train": X_train,
            "X_test": X_test,
            "y_train": y_train,
            "y_test": y_test,
            "X_train_procesado": X_train_procesado,
            "X_test_procesado": X_test_procesado
        }
```

```python
agente_1 = AgenteNormalizador()

resultado_agente_1 = agente_1.preparar_datos(df)

print("Agente 1 ejecutado correctamente")
print(
    "Dataset limpio:",
    resultado_agente_1["df_limpio"].shape
)
print(
    "Entrenamiento procesado:",
    resultado_agente_1["X_train_procesado"].shape
)
print(
    "Prueba procesada:",
    resultado_agente_1["X_test_procesado"].shape
)
```

```python
#Agente 2 Entrenador. Entrenará varios modelos, evaluar accuracy, precision, recall,F1 y AUC
#Seleccionará automaticamente el mejor jjsjs
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

import pandas as pd


class AgenteEntrenador:
    def __init__(self):
        self.modelos = {
            "Regresión Logística": LogisticRegression(
                max_iter=1000,
                random_state=42
            ),

            "Árbol de Decisión": DecisionTreeClassifier(
                max_depth=10,
                random_state=42
            ),

            "Random Forest": RandomForestClassifier(
                n_estimators=100,
                random_state=42,
                n_jobs=-1
            )
        }

        self.resultados = []
        self.mejor_modelo = None
        self.nombre_mejor_modelo = None

    def entrenar_y_evaluar(
        self,
        X_train,
        X_test,
        y_train,
        y_test
    ):
        self.resultados = []

        for nombre, modelo in self.modelos.items():

            print(f"Entrenando: {nombre}")

            modelo.fit(X_train, y_train)

            predicciones = modelo.predict(X_test)

            probabilidades = modelo.predict_proba(X_test)[:, 1]

            resultado = {
                "Modelo": nombre,
                "Accuracy": accuracy_score(
                    y_test,
                    predicciones
                ),
                "Precision": precision_score(
                    y_test,
                    predicciones,
                    zero_division=0
                ),
                "Recall": recall_score(
                    y_test,
                    predicciones,
                    zero_division=0
                ),
                "F1 Score": f1_score(
                    y_test,
                    predicciones,
                    zero_division=0
                ),
                "AUC": roc_auc_score(
                    y_test,
                    probabilidades
                )
            }

            self.resultados.append(resultado)

        tabla_resultados = pd.DataFrame(self.resultados)

        tabla_resultados = tabla_resultados.sort_values(
            by="F1 Score",
            ascending=False
        ).reset_index(drop=True)

        self.nombre_mejor_modelo = tabla_resultados.loc[
            0,
            "Modelo"
        ]

        self.mejor_modelo = self.modelos[
            self.nombre_mejor_modelo
        ]

        return tabla_resultados
```

```python
#Probamos el agente usando la salida del Agente 1
agente_2 = AgenteEntrenador()

tabla_resultados = agente_2.entrenar_y_evaluar(
    resultado_agente_1["X_train_procesado"],
    resultado_agente_1["X_test_procesado"],
    resultado_agente_1["y_train"],
    resultado_agente_1["y_test"]
)

print("\nComparación de modelos:")
display(tabla_resultados)

print(
    "\nMejor modelo:",
    agente_2.nombre_mejor_modelo
)
```
