```python
import pandas as pd
import numpy as np

df = pd.read_csv("hotel_bookings.csv")

df.head()

```

```python
# Tamaño del dataset
filas, columnas = df.shape

print("Cantidad de filas:", filas)
print("Cantidad de columnas:", columnas)
print("Cantidad total de celdas:", filas * columnas)

# Primeras filas
display(df.head())

# Información de las columnas
df.info()

# Valores faltantes
faltantes = pd.DataFrame({
    "Cantidad": df.isnull().sum(),
    "Porcentaje": (df.isnull().sum() / len(df)) * 100
})

faltantes = faltantes[faltantes["Cantidad"] > 0]
display(faltantes.sort_values("Cantidad", ascending=False))

# Duplicados
print("Filas duplicadas:", df.duplicated().sum())
```

```python
df_original = df.copy()
df_limpio = df.copy()

print("Copia creada correctamente")
print("Filas originales:", df_limpio.shape[0])
print("Columnas originales:", df_limpio.shape[1])
```

```python
duplicados_antes = df_limpio.duplicated().sum()

df_limpio = df_limpio.drop_duplicates().reset_index(drop=True)

print("Duplicados encontrados:", duplicados_antes)
print("Filas después de eliminar duplicados:", df_limpio.shape[0])
```

```python
#Eliminar la columna "company" ya que tiene mas del 94% de valores faltantes, imputarala no seria confiable
df_limpio = df_limpio.drop(columns=["company"])

print("Columna 'company' eliminada")
print("Cantidad de columnas:", df_limpio.shape[1])
```

```python
#Tratar los valores faltantes
#Para children, usamos la mediana
df_limpio["children"] = df_limpio["children"].fillna(
    df_limpio["children"].median()
)
#Para country usamos, usamos la moda
df_limpio["country"] = df_limpio["country"].fillna(
    df_limpio["country"].mode()[0]
)
#Para agent, conviene usar 0, porque generalmente indica que la reserva no fue gestionada por un agente
df_limpio["agent"] = df_limpio["agent"].fillna(0)
#Comprobar el resultado
print("Valores faltantes restantes:")
print(df_limpio.isnull().sum()[df_limpio.isnull().sum() > 0])

print("\nDimensiones actuales:")
print(df_limpio.shape)

resumen_limpieza = pd.DataFrame({
    "Tipo de dato": df_limpio.dtypes,
    "Valores faltantes": df_limpio.isnull().sum()
})

display(resumen_limpieza)
```

```python
print(df_limpio["reservation_status"].value_counts())

display(
    df_limpio[
        ["is_canceled", "reservation_status", "reservation_status_date"]
    ].head(15)
)
```

```python
#Eliminacion de columnas con fugas de informacion
columnas_fuga = [
    "reservation_status",
    "reservation_status_date"
]

df_limpio = df_limpio.drop(columns=columnas_fuga)

print("Columnas eliminadas:", columnas_fuga)
print("Dimensiones actuales:", df_limpio.shape)
```

```python
reservas_sin_huespedes = df_limpio[
    (df_limpio["adults"] == 0) &
    (df_limpio["children"] == 0) &
    (df_limpio["babies"] == 0)
]

print("Reservas sin huéspedes:", len(reservas_sin_huespedes))

#Eliminar esas reservas sin huespedes
df_limpio = df_limpio[
    ~(
        (df_limpio["adults"] == 0) &
        (df_limpio["children"] == 0) &
        (df_limpio["babies"] == 0)
    )
].reset_index(drop=True)

print("Dimensiones después de eliminar reservas inválidas:")
print(df_limpio.shape)

#Recisamos el balance de la variable objetivo, saber cuantas reservas fueron canceladas
#y cuantas reservas no fueron canceladas
conteo_cancelaciones = df_limpio["is_canceled"].value_counts()

porcentaje_cancelaciones = (
    df_limpio["is_canceled"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

print("Cantidad:")
print(conteo_cancelaciones)

print("\nPorcentaje:")
print(porcentaje_cancelaciones)
print(df_limpio.shape)
print(df_limpio["is_canceled"].value_counts())
```
